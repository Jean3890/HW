```python
from collections import UserList
from math import pi


#створюємо батьківский клас
class GeometricFigure():
    figure_name = ''
    area = 0

    def __init__(self, figure_name):
        self.figure_name = figure_name

    def calculating_the_area(self):
        raise NotImplementedError("Please Implement this method")

#створюємо дочірній клас квадрат
class Square(GeometricFigure):
    # def __init__(self, side = 0):
    #     self._side = side
    
    def set_side(self, x):
        self._side = x

    def calculating_the_area(self):
        self.area = self._side * self._side

#створюємо дочірній клас коло
class Circle(GeometricFigure):
    radius = 0
    
    def set_radius(self, radius):
        self.radius = radius

    def calculating_the_area(self):
        self.area = self.radius * self.radius * round(pi,2)

#клас підрахунку площі всіх фігур
class Calculation(UserList):
    def calculate_total_area(self):
        total_area = 0
        delimeter = '-' * 48
        print("|{:<20}|{:>25}|".format('Figure', 'Area'))
        print(delimeter)

        for figure in self.data:
            figure.calculating_the_area()
            total_area += figure.area
            print("|{:<20}|{:>25}|".format(figure.figure_name, figure.area))

        print(delimeter)
        print("|{:<19} |{:>25}|".format('Total Area', round(total_area,2)))



result = Calculation()

#створюємо 4 квадрати
square_1 = Square('Square 1')
square_1.set_side(3)

square_2 = Square('Square 2')
square_2.set_side(2)

square_3 = Square('Square 3')
square_3.set_side(6)

square_4 = Square('Square 4')
square_4.set_side(10)

#створюємо 4 кола
circle_1 = Circle('Circle 1')
circle_1.set_radius(5)

circle_2 = Circle('Circle 2')
circle_2.set_radius(4)

circle_3 = Circle('Circle 3')
circle_3.set_radius(7)

circle_4 = Circle('Circle 4')
circle_4.set_radius(6)

result.append(square_1)
result.append(square_2)
result.append(square_3)
result.append(square_4)
result.append(circle_1)
result.append(circle_2)
result.append(circle_3)
result.append(circle_4)

result.calculate_total_area()
```
